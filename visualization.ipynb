```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("master.csv")
```

```python
#albania data
albania=df[:263]
albania.head(10)
```

```python
#age and suicide numbers
sns.set(style="whitegrid")
g=sns.catplot(x="age",y="suicides_no",hue="sex", kind="bar", data=albania)
plt.xticks(rotation=70)
```

```python
#box plots for age adn suicide number
b=sns.boxplot(x="year", y=albania.population/10**6, data=albania)
plt.xticks(rotation=70)
```

```python
s=sns.lineplot(x="year", y=albania.population/10**6,data=albania)
plt.xticks(rotation=70)
```

```python
sns.set(style="whitegrid")
```

```python
sns.violinplot(x="sex",y="suicides_no",data=albania, split=True,inner="quart")
plt.xticks(rotation=70)
```
